```python
ages=df['Age'].dropna(how='any')
```

```python
ages=list(ages)
ages=set(ages)
```

```python
from collections import Counter
```

```python
age_counter=Counter()
for age in ages:
    age_counter.update(str(age).split('.'))
```

```python
age_counter.pop('0')
```

```python
underage=['18']
for item in underage:
        age_counter.pop(item)
```
